```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('data/2024.csv')
```

```python
data.columns = ['State/UT', 'District', 'Disease', 'No_of_Cases', 'No_of_Deaths', 'Date_of_Start', 'Current_Status']
```

```python
new_data=data[['State/UT', 'District', 'Disease', 'No_of_Cases','Date_of_Start']]
new_data['Date_of_Start'] = pd.to_datetime(new_data['Date_of_Start'])
```

```python
new_data.head()
```

```python
state_df=new_data['State/UT'].unique()
district_df=new_data['District'].unique()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = data

# 1. Total Cases per Disease
total_cases_per_disease = df.groupby("Disease")["No_of_Cases"].sum()

# 2. Total Cases per State/UT
total_cases_per_state = df.groupby("State/UT")["No_of_Cases"].sum()

# 3. Average Number of Cases per Disease
average_cases_per_disease = df.groupby("Disease")["No_of_Cases"].mean()

# 4. Bar chart for total cases per State/UT
plt.figure(figsize=(10,6))
sns.barplot(x=total_cases_per_state.index, y=total_cases_per_state.values, palette='viridis')
plt.title('Total Cases per State/UT')
plt.xlabel('State/UT')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 5. Bar chart for total cases per Disease
plt.figure(figsize=(10,6))
sns.barplot(x=total_cases_per_disease.index, y=total_cases_per_disease.values, palette='plasma')
plt.title('Total Cases per Disease')
plt.xlabel('Disease')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 6. Pie chart for Disease distribution
plt.figure(figsize=(8,8))
plt.pie(total_cases_per_disease, labels=total_cases_per_disease.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Disease Distribution by Cases')
plt.show()

# 7. Summary statistics
summary_stats = df.describe()

total_cases_per_disease, total_cases_per_state, average_cases_per_disease, summary_stats
```
